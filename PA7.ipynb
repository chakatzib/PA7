```python
import math
class calculator():
    def __init__(self,input1,input2):
        self.input1=input1
        self.input2=input2
    def addition(self):
        return self.input1+self.input2
    def subtraction(self):
        return self.input1-self.input2
    def multiplication(self):
        return self.input1*self.input2
    def division(self):
        return self.input1/self.input2
    def exponent(self):
        return self.input1**self.input2
    def sine(self):
        return math.sin(self.input1)
    def cosine(self):
        return math.cos(self.input1)
    def tangent(self):
        return math.tan(self.input1)
    def arcsine(self):
        return math.asin(self.input1)
    def arccosine(self):
        return math.acos(self.input1)
    def arctangent(self):
        return math.atan(self.input1)
    def inversehyperbolicsine(self):
        return math.asinh(self.input1)
    def inversehyperboliccosine(self):
        return math.acosh(self.input1)
    def inversehyperbolictangent(self):
        return math.atanh(self.input1)
    def squareroot(self):
        return math.sqrt(self.input1)
input_1=int(input("Enter first number: "))
input_2=int(input("Enter second number: "))
my_calc=calculator(input_1,input_2)

choice=1
while choice!=0:
    print()
    print("Options marked with a * indicate that only one number is used. The first number takes priority over the second number.")
    print("0. Exit")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    print("5. Exponent/Power")
    print("6. Sine*")
    print("7. Cosine*")
    print("8. Tangent*")
    print("9. Arc Sine(number from -1 to 1)*")
    print("10. Arc Cosine(number from -1 to 1)*")
    print("11. Arc Tangent*")
    print("12. Inverse Hyperbolic Sine*")
    print("13. Inverse Hyperbolic Cosine(must be positive)*")
    print("14. Inverse Hyperbolic Tangent(must be from -.99 to .99)*")
    print("15. Square root*")
    choice=int(input("Enter your choice...."))
    
    if choice==1:
        print(f"Result: {my_calc.addition()}")
    elif choice==2:
        print(f"Result: {my_calc.subtraction()}")
    elif choice==3:
        print(f"Result: {my_calc.multiplication()}")
    elif choice==4:
        print(f"Result: {my_calc.division()}")
    elif choice==5:
        print(f"Result: {my_calc.exponent()}")
    elif choice==6:
        print(f"Result: {my_calc.sine()}")
    elif choice==7:
        print(f"Result: {my_calc.cosine()}")
    elif choice==8:
        print(f"Result: {my_calc.tangent()}")
    elif choice==9:
        print(f"Result: {my_calc.arcsine()}")
    elif choice==10:
        print(f"Result: {my_calc.arccosine()}")
    elif choice==11:
        print(f"Result: {my_calc.arctangent()}")
    elif choice==12:
        print(f"Result: {my_calc.inversehyperbolicsine()}")
    elif choice==13:
        print(f"Result: {my_calc.inversehyperboliccosine()}")
    elif choice==14:
        print(f"Result: {my_calc.inversehyperbolictangent()}")
    elif choice==15:
        print(f"Result: {my_calc.squareroot()}")
    elif choice==0:
        print("Exit")
    else: 
        print("Sorry, invalid choice. Read the rules, kindly adjust.")
        
```
